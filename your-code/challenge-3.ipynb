```python
Sub Problem 1: How to aggregate the amount_spent for unique customers?
```

```python
agg_amount_spent = orders.groupby('CustomerID')['amount_spent'].sum().reset_index()

agg_amount_spent
```

```python
Sub Problem 2: How to select customers whose aggregated amount_spent is in a given quantile range?
```

```python
lower_quantile =0.25  
upper_quantile = 0.75  


lower_threshold = orders['amount_spent'].quantile(lower_quantile)
upper_threshold = orders['amount_spent'].quantile(upper_quantile)

upper_threshold

```

```python
Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?
```

```python
lower_quantile = 0.25
upper_quantile = 0.75


vip_customers = orders[orders['amount_spent'] > orders['amount_spent'].quantile(upper_quantile)]['CustomerID']


preferred_customers = orders[(orders['amount_spent'] > orders['amount_spent'].quantile(lower_quantile)) &
                           (orders['amount_spent'] <= orders['amount_spent'].quantile(upper_quantile))]['CustomerID']


display("VIP Customers:", vip_customers)
display("Preferred Customers:", preferred_customers)







```

## Q2: How to identify which country has the most VIP Customers?

```python
vip_customers= orders.groupby(['Country','amount_spent'])['CustomerID'].nunique()
vip_customers = vip_customers.reset_index()
vip_customers
```

```python
preferred_customers= orders.groupby(['Country','amount_spent'])['CustomerID'].nunique()
preferred_customers = preferred_customers.reset_index()
preferred_customers
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
country_comb = pd.concat([vip_customers, preferred_customers], axis=0)
```
